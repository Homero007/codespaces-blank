```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt

# Asignar colores para los rectngulos (BGR)
COLORS = {
    'red': [0, 0, 220],          # machos adultos
    'magenta': [255, 100, 255],  # machos subadultos
    'brown': [30, 80, 120],      # hembras adultas
    'blue': [230, 80, 60],       # juveniles
    'green': [60, 180, 80]       # cachorros
}

def load_images(image_path, prototype_path):
    """Carga las imgenes y el prototipo usando OpenCV."""
    image = cv2.imread(image_path)
    prototype = cv2.imread(prototype_path, cv2.IMREAD_GRAYSCALE)
    return image, prototype

def process_image(image, prototype, threshold):
    """Procesa la imagen para encontrar coincidencias usando correlacin cruzada."""
    result = cv2.matchTemplate(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY), prototype, cv2.TM_CCOEFF_NORMED)
    return np.where(result >= threshold)

def draw_rectangles(image_result, loc, prototype_shape, color):
    """Dibuja los cuadros envolventes en las ubicaciones encontradas."""
    for pt in zip(*loc[::-1]):
        cv2.rectangle(image_result, pt, (pt[0] + prototype_shape[1], pt[1] + prototype_shape[0]), COLORS[color], 2)

def save_and_show_image(image_result, output_path):
    """Guarda y muestra la imagen con los cuadros envolventes."""
    cv2.imwrite(output_path, image_result)
    plt.imshow(cv2.cvtColor(image_result, cv2.COLOR_BGR2RGB))
    plt.title('Imagen con Cuadros Envolventes')
    #plt.show()

def object_identification(image_path, prototype_path, image_result, color, threshold, output_path='image_resultado.jpg'):
    """Realiza la identificacin de objetos y dibuja cuadros envolventes."""
    image, prototype = load_images(image_path, prototype_path)
    loc = process_image(image, prototype, threshold)
    draw_rectangles(image_result, loc, prototype.shape, color)
    save_and_show_image(image_result, output_path)
    return image_result

def main():
    # Variables de inicio
    image_path = '/content/41.jpg'
    prototype_folder = '/content/plantillas'
    output_path = '/content/imagen_resultado.jpg'
    threshold = 0.70

    # Inicializar el resultado con la imagen original
    image_result = cv2.imread(image_path).copy()

    # Organizar los prototipos por color/clase
    classes = {key: [] for key in COLORS.keys()}
    paths = os.listdir(prototype_folder)

    for filename in paths:
        color, _ = filename.split('-')
        classes[color].append(filename)

    # Procesar cada prototipo
    for color, file_paths in classes.items():
        for file in file_paths:
            image_result = object_identification(
                image_path,
                os.path.join(prototype_folder, file),
                image_result,
                color,
                threshold,
                output_path
            )

if __name__ == "__main__":
    main()
```
